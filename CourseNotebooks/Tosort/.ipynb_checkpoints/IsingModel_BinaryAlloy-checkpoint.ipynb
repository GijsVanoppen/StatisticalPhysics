```python
## Python libraries to load
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation

import numpy as np
from tqdm import tqdm
import copy
```

## 1. Support functions ##
Functions used to generate look-up tables, calculated specific properties and perform single MMC steps.

```python
## Generating look-up tables
def get_PBC(L:int)->tuple:
    """
    Create two look-up tables for the +1 and -1 sites, 
    taking periodic boundary conditions into account.
    
    parameters:
        - L : linear dimension of the spin-lattice
    """
    kp=np.zeros(L,dtype=int)
    km=np.zeros(L,dtype=int)
    for i in range(L):
        kp[i]=i+1
        km[i]=i-1
    kp[L-1]=0
    km[0]=L-1
    return kp,km
    
def get_transProb(Jc,H,K,kT)->np.ndarray:
    """
    Create a look-up table with the possible calculated transition probabilities "w".
    For the binary alloy, the transitions will be an exchange of two sites. The two sites
    can be anywhere in the system (not-NN).
    
    parameters:
        - Jc  : NN-interactions.
        - H  : chemical potential term
        - K  : Bias term
        - kT : Thermal factor kB*T (kB in meV/K)
    """
    delta=np.zeros((3,9))
    w=np.zeros((3,9))
    for i in [-1,1]:
        for j in [-4,-2,0,2,4]:
            delta[i,j]=2.0*i*(j*Jc+H)
    for i in [-1,1]:
        for j in [-4,-2,0,2,4]:
            w[i,j]=1
            if (delta[i,j]>0):
                w[i,j]=np.exp(-delta[i,j]/kT)
    return w

def get_NNborder()->np.ndarray:
    """
    Create a lookup table to store the number of 'border' positions.
    
    """
    delta=np.zeros((3,9))
    delta[1,4]=0
    delta[1,2]=1
    delta[1,0]=2
    delta[1,-2]=3
    delta[1,-4]=4
    delta[-1,4]=4
    delta[-1,2]=3
    delta[-1,0]=2
    delta[-1,-2]=1
    delta[-1,-4]=0
    return delta
    

def get_transProb_NN(Jc,H,K,kT)->np.ndarray:
    """
    Create a look-up table with the possible calculated transition probabilities "w".
    For the binary alloy, the transitions will be an exchange of two sites. This function
    provides the transition probabilities in case of the special situation 
    of two NN sites being exchanged.
    
    parameters:
        - Jc  : NN-interactions.
        - H  : chemical potential term
        - K  : Bias term
        - kT : Thermal factor kB*T (kB in meV/K)
    """
    delta=np.zeros((3,9))
    w=np.zeros((3,9))
    i=1
    for j in [-4,-2,0,2]:
        delta[i,j]=((j+1)*Jc+0.75*H)
    i=-1
    for j in [-2,0,2,4]:
        delta[i,j]=(-(j-1)*Jc-0.75*H)
    for i in [-1,1]:
        for j in [-4,-2,0,2,4]:
            w[i,j]=1
            if (delta[i,j]>0):
                w[i,j]=np.exp(-2.0*delta[i,j]/kT)
    return w
###################################################################
## grid-functions
def init_sampleLR(L)->np.ndarray:
    """
    Generate a grid with two large domains:
        - L  : linear dimension of the spin-lattice.
    """
    grid=np.zeros((L,L))
    grid+=1
    if (L%2)==0 : #even
        L2=int(L*0.5)
        grid[0:L2,:]=-1
    else:
        L2=int((L-1)*0.5)
        grid[0:L2,:]=-1
        grid[L2,0:L2]=-1 # half-1 of last line flipped as well

    return grid

def init_sampleRand(L)->np.ndarray:
    """
    Generate a grid in a (uniform) random noise configuration
    parameters:
        - L  : linear dimension of the spin-lattice.
    """
    grid=np.zeros((L,L))
    grid+=1
    #select L²/2 positions to be -1
    NR=np.arange(L*L)    #create a linear list of indices
    np.random.shuffle(NR)#shuffle this list
    for idx in range(int(L*L*0.5)):#only half need to be flipped
        j=(NR[idx] % L)  #mod gives value from 0..L-1
        i=int((NR[idx]-j)/L)
        grid[i,j]=-1
    return grid


## MC functions
def do_mcs_step_RandSweep(N,g,L,iplus,imin,w,wNN):
    """
    Perform a N single Monte-Carlo exchange events. 
    
    parameters:
        - N : number of exchanges to perform
        - g : 2D spin lattice
        - L : linear dimension of the lattice.
        - iplus, imin : pregenerated look-up tables for the +1 and -1 sites.
        - w : look-up table with transition probabilities.
        - wNN : look-up table with transition probabilities for NN exchanges.
    """
    ##generate all possibly required random numbers in one go (less overhead)
    rn=np.random.rand(N) #uniform over 0..1
    L2=L**2
    ngrid=np.arange(L2).reshape(L,L)        # make a mapping of the 2D grid to a linear value
    sitesA=np.where(g>0,ngrid,-5).flatten() # find the A sites
    sitesB=np.where(g<0,ngrid,-5).flatten() # find the B sites
    sitesA=sitesA[sitesA != -5]  # and now just remove the -5's
    sitesB=sitesB[sitesB != -5]  # and now just remove the -5's
    LA=len(sitesA)
    LB=len(sitesB)
    cnt=0
    for k in range(N):
        sA=np.random.choice(sitesA,size=1) # randomly select an A and B site
        sB=np.random.choice(sitesB,size=1)
        #transform grid-index-number to grid-position...rows are first index, so are L*1..n, second index are colums--> mod L
        jA=(sA % L)  #mod gives value from 0..L-1
        iA=int((sA-jA)/L)
        jB=(sB % L)  #mod gives value from 0..L-1
        iB=int((sB-jB)/L)
        
        siteA=g[iA,jA]
        somA=g[iplus[iA],jA]+g[imin[iA],jA]+g[iA,iplus[jA]]+g[iA,imin[jA]]
        k1A=int(somA)
        k2A=int(siteA)
        siteB=g[iB,jB]
        somB=g[iplus[iB],jB]+g[imin[iB],jB]+g[iB,iplus[jB]]+g[iB,imin[jB]]
        k1B=int(somB)
        k2B=int(siteB)
        #check if sites are NN...linear L1 distance is 1 for NN... 
        dist=(abs(iA-iB)%(L-1))+(abs(jA-jB)%(L-1))  # the %L-1 puts the PBC edge-distance to zero puts L1 distance to 0 for NN
        if (dist<2):#the sites are NN      
            weight=wNN[k2A,k1A]*wNN[k2B,k1B]
        else:
            weight=w[k2A,k1A]*w[k2B,k1B]
        
        if (rn[k]<weight):#flip
                g[iA,jA]=-g[iA,jA]
                g[iB,jB]=-g[iB,jB]
                #we also need to switch these in our arrays keeping track of A and B sites 
                #(this is the reason we can not generate a random sequence in advance)
                sitesA=np.where(sitesA==sA,sB,sitesA) # note that each value in the sitesA and B arrays is unique :-)
                sitesB=np.where(sitesB==sB,sA,sitesB)
                
## Collect/calculate properties
def calc_TotB(g: np.ndarray, L, iplus, imin, dNN )-> float:
    """ 
    Calculate the number of edge atoms:
        - g  : 2D grid of lattice sites   
    """
    cnt=0
    for i in range(L):
        for j in range(L):
            site=int(g[i,j])
            som=int(g[iplus[i],j]+g[imin[i],j]+g[i,iplus[j]]+g[i,imin[j]])
            cnt+=dNN[site,som]
    
    return cnt*0.5

def calc_TotE(g: np.ndarray,L,Jc,H,K,z,iplus,imin)-> float:
    """ 
    Calculate the total energy of the 2D Ising model using global J & B
    parameters:
        - g  : 2D grid of lattice sites
    """
    TE = K*4.0*L*L
    HonZ=1.0*H/z
    for i in range(L):
        for j in range(L):
            som=int(g[iplus[i],j]+g[imin[i],j]+g[i,iplus[j]]+g[i,imin[j]])
            TE-=Jc*g[i,j]*som + HonZ*(4*g[i,j]+som)
    
    return 0.5*TE #return the total energy
```

## 2. 2D Ising model for a binary alloy ##
The alloy consists of two types of atoms A and B. The interactionenergies between atom-pairs is given by $-\epsilon_{AA}$, $-\epsilon_{BB}$, and $-\epsilon_{AB}$ for A-A, B-B, and A-B, respectively. Note the minus sign, indicating energies to be negative (for stable configurations).

The interaction energies are not directly implemented into the Ising model, instead they are linked to three general factors:
 - **Jc** : The coupling between atoms (irrespective of the atomic species)
 - **H**  : A chemical potential like factor modifying the "fraction density"
 - **K**  : A bias factor
 
The interaction energie for a site-pair in the ising model is given by:<br />
$\epsilon_{i,j}=-Jc\sigma_i\sigma_j - \frac{1}{z}H(\sigma_i+\sigma_j) + K$<br />
Solving the system of equatioins obtained for the three possible pairs: AA, AB, and BB the constants Jc, H, and K are found to be<br />
 - $Jc= 0.25*(\epsilon_{AA}+\epsilon_{BB}-2\epsilon_{AB})$
 - $H = 0.25*z*(\epsilon_{AA}-\epsilon_{BB})$
 - $K =-0.25*(\epsilon_{AA}+\epsilon_{BB}+2\epsilon_{AB})$
In this $z$ is the number of nearest neighbours (*i.e.*, 4 in a 2D square lattice).
    
    
### 2.1. Simulation initialisation ###

```python
## Provide the interaction energies
eAA = 150.0   # interaction energie is -50 meV
eBB = 100.0
eAB = 15.0

## from these the model constants are calculated:
z      = 4.0                        # The number of nearest neighbours
Jc     = 0.25*(eAA+eBB-2.0*eAB)     # The coupling constant in energy "meV" (1k_BT = 26meV@RT)
H      = 0.25*z*(eAA-eBB)
K      = -0.25*(eAA+eBB+2.0*eAB)

T      = 3.0        # Temperature in Kelvin
kBT    = T*8.617333e-2# k_B in meV/K

latdim = 50           # Lateral lattice dimension
N      = latdim**2    # Total number of lattice sites in a 2D lattice

print("Jc = ",Jc," meV")
print("H  = ",H ," meV")
print("K  = ",K ," meV")
```

### 2.2. Simulation: main loop ###

```python
## The main loop
mcsmax  = 5000 # Number of lattice sweeps to perform
Ninit   = 0    # Number of sweeps used as initialisation of the system
Nsample = 100   # System properties are collected every # systems-sweeps

## Create a gif animation of the experiment
fig = plt.figure(figsize=(6,6))
x = np.arange(0, latdim)
y = np.arange(0, latdim).reshape(-1, 1)
frames = list()

## Properties we will be tracking
propE = list() # The total energy of the system
propB = list() # The total number of edge ions in the system
avgE  = list() # Running averages...
avgB  = list()
cntp  = 0      # Counter of property-indices

## Look-up tables
kp, km = get_PBC(latdim)
w = get_transProb(Jc,H,K,kBT)
wNN = get_transProb_NN(Jc,H,K,kBT)
dNN = get_NNborder()
#print("trans1=",w)
#print("trans2=",wNN)

## Initialisation
grid = init_sampleRand(latdim)
#grid = init_sampleLR(latdim)

## Monte Carlo run
cnt=0
xx=list()
bar = tqdm(total=mcsmax)
bar.update(Ninit)
for mcs in range(mcsmax):
    do_mcs_step_RandSweep(Nsample,grid,latdim,kp,km,w,wNN)
    #do_mcs_step_RandSweep(grid,latdim,kp,km,w)# do a single update-round
    if mcs > Ninit:
        cnt += 1
        if cnt >= Nsample:
            propE.append(calc_TotE(grid,latdim,Jc,H,K,z,kp,km))
            propB.append(calc_TotB(grid, latdim, kp, km, dNN ))
            avgE.append(np.mean(propE[-10:-1]))
            avgB.append(np.mean(propB[-10:-1]))
            xx.append(mcs)
            cntp += 1
            cnt   = 0
            
            title="Iteration "+str(mcs)+" /"+str(mcsmax)
            ttl = plt.text(0.5*latdim, latdim*1.03, title, horizontalalignment='center', verticalalignment='top')
            frames.append((plt.pcolor(x, y,grid[:,:] , cmap="bwr", 
                              norm=plt.Normalize(0, 1),label=title), ttl, ))
            
            bar.update(Nsample)
bar.close()
```

### 2.2. Plot the results ###

```python
%matplotlib notebook
## Gif animation
im_ani = animation.ArtistAnimation(fig, frames, interval=100, repeat_delay=10,
                                   blit=False)
# To save this second animation with some metadata, use the following command:
im_ani.save('Ising_Alloy.gif', dpi=40, writer='imagemagick', fps = 60, bitrate = 1500, 
            metadata={'artist':'https://DannyVanpoucke.be'})

plt.show()
```

```python
## The energy
%matplotlib inline 
plt.figure(figsize=(16,6))

plt.plot(xx, propE, color="red"   ,linewidth=1, zorder=-1, alpha=1)
plt.plot(xx, avgE,  color="blue"  ,linewidth=1, zorder=-1, alpha=1)
plt.xscale("linear")
plt.yscale("linear")
ddx=0.01
maxE=max([max(propE)*(1.0+ddx),max(propE)*(1.0-ddx)])
minE=min([min(propE)*(1.0+ddx),min(propE)*(1.0-ddx)])
plt.axis([0,mcsmax,minE,maxE])
Xlst=np.linspace(0,mcsmax,11)
plt.xticks(Xlst,rotation=45,fontsize=18)
Ylst=(np.linspace(0,6,6))*(abs(maxE-minE)*0.2)+minE
plt.yticks(Ylst,fontsize=18)

plt.xlabel("# simulation step",fontsize=22,fontweight="bold")
plt.ylabel("E per site (meV)",fontsize=22,fontweight="bold")
plt.grid(b=True, which='both', axis='both', color='lightgrey', linestyle='--', linewidth=1)
plt.show()
```

## 3. Temperature dependence in Binary Alloys ##
Using the above setup, we could now investigate the phase transition between a perfect alloy and phase-segregation as function of temperature.
Due to the randomness of the system (initial setup, and updates), an ensemble of models is run for every temperature. The parameters to track are the total energy and border-size. In addition we, also keep track of the "convergence" of both parameters at the end of a fixed number of update runs.

```python
## Initial settings
eAA = 150.0   # interaction energie is -50 meV
eBB = 100.0
eAB = 15.0

## from these the model constants are calculated:
z      = 4.0                        # The number of nearest neighbours
Jc     = 0.25*(eAA+eBB-2.0*eAB)     # The coupling constant in energy "meV" (1k_BT = 26meV@RT)
H      = 0.25*z*(eAA-eBB)
K      = -0.25*(eAA+eBB+2.0*eAB)

nT     = 150
scale  = 25.0
Tlst   = np.arange(start=1,stop=1+nT, step=1)*scale # temperatuur in Kelvin
kBTlst = Tlst*8.617333e-2# k_B in meV/K

latdim = 30           # Lateral lattice dimension
N      = latdim**2    # Total number of lattice sites in a 2D lattice
nMods  = 20           # number of models in the ensemble

print("Jc = ",Jc," meV")
print("H  = ",H ," meV")
print("K  = ",K ," meV")
```

```python
## The main loop
mcsmax  = 10000 # Number of lattice sweeps to perform
Ninit   = 9500    # Number of sweeps used as initialisation of the system
Nsample = 100   # System properties are collected every # systems-sweeps

## Properties we will be tracking
propE = np.zeros((nT,nMods)) # The total energy of the system
propB = np.zeros((nT,nMods)) # The total number of edge ions in the system
convE = np.zeros((nT,nMods)) # convergence...
convB = np.zeros((nT,nMods))

## Look-up tables
kp, km = get_PBC(latdim)
dNN = get_NNborder()
perSite = 1.0/(latdim**2)
frac = perSite*100.0
## Progressbar
bar = tqdm(total=nT)
bar.update()
for iT in range(nT):
    kBT=kBTlst[iT]
    ## Look-up tables
    w = get_transProb(Jc,H,K,kBT)
    wNN = get_transProb_NN(Jc,H,K,kBT)
    
    for im in range(nMods): 
        ## Initialisation
        grid = init_sampleRand(latdim)
        #grid = init_sampleLR(latdim)
        do_mcs_step_RandSweep(Ninit,grid,latdim,kp,km,w,wNN) # as we run fixed # updates, 
                                                             # we only need to track changes near the end
        mcs=Ninit
        pE=list()
        pB=list()
        cnt=0
        while mcs < mcsmax: # remainder of the updates
            do_mcs_step_RandSweep(Nsample,grid,latdim,kp,km,w,wNN)
            mcs+=Nsample
            cnt+=1
            pE.append(calc_TotE(grid,latdim,Jc,H,K,z,kp,km)*perSite)
            pB.append(calc_TotB(grid, latdim, kp, km, dNN )*frac)
        propE[iT,im]=pE[-1]
        propB[iT,im]=pB[-1]
        convE[iT,im]=np.mean(list(pE[i+1]-pE[i] for i in range(cnt-1)))
        convB[iT,im]=np.mean(list(pB[i+1]-pB[i] for i in range(cnt-1)))
            
    bar.update()
bar.close()
```

### 3.2. Plot Results ###


```python
# The energy
plt.figure(figsize=(16,6))
plt.plot(Tlst, np.mean(propE,axis=1), color="red"   ,linewidth=1, zorder=1, alpha=1, label="Average Value")
plt.fill_between(Tlst, np.mean(propE,axis=1)-2*np.std(propE,axis=1), np.mean(propE,axis=1)+2*np.std(propE,axis=1), color="grey",
                  zorder=2, alpha=0.25)
plt.fill_between(Tlst, np.mean(propE,axis=1)-2*np.mean(convE,axis=1), np.mean(propE,axis=1)+2*np.mean(convE,axis=1), color="green",
                  zorder=0, alpha=0.25)

plt.xscale("linear")
plt.yscale("linear")
ddx=0.01
maxE=np.amax([np.amax(propE)*(1.0+ddx),np.amax(propE)*(1.0-ddx)])
minE=np.amin([np.amin(propE)*(1.0+ddx),np.amin(propE)*(1.0-ddx)])
plt.axis([0,nT*scale,minE,maxE])
Xlst=np.linspace(0,nT*scale,11)
plt.xticks(Xlst,rotation=45,fontsize=18)
Ylst=(np.linspace(0,6,6))*(abs(maxE-minE)*0.2)+minE
plt.yticks(Ylst,fontsize=18)

plt.xlabel("# temperature [K]",fontsize=22,fontweight="bold")
plt.ylabel("E per site (meV)",fontsize=22,fontweight="bold")
plt.grid(b=True, which='both', axis='both', color='lightgrey', linestyle='--', linewidth=1)
plt.legend()
plt.show()
```

```python
# The borders
plt.figure(figsize=(16,6))
plt.plot(Tlst, np.mean(propB,axis=1), color="red"   ,linewidth=1, zorder=1, alpha=1, label="Average Value")
plt.fill_between(Tlst, np.mean(propB,axis=1)-2*np.std(propB,axis=1), np.mean(propB,axis=1)+2*np.std(propB,axis=1), color="grey",
                  zorder=-1, alpha=0.25)
plt.fill_between(Tlst, np.mean(propB,axis=1)-2*np.mean(convB,axis=1), np.mean(propB,axis=1)+2*np.mean(convB,axis=1), color="green",
                  zorder=0, alpha=0.25)

plt.xscale("linear")
plt.yscale("linear")
ddx=0.01
maxB=np.amax([np.amax(propB)*(1.0+ddx),np.amax(propB)*(1.0-ddx)])
minB=np.amin([np.amin(propB)*(1.0+ddx),np.amin(propB)*(1.0-ddx)])
plt.axis([0,nT*scale,minB,maxB])
Xlst=np.linspace(0,nT*scale,11)
plt.xticks(Xlst,rotation=45,fontsize=18)
Ylst=(np.linspace(0,6,6))*(abs(maxB-minB)*0.2)+minB
plt.yticks(Ylst,fontsize=18)

plt.xlabel("# temperature [K]",fontsize=22,fontweight="bold")
plt.ylabel("fraction of edge sites",fontsize=22,fontweight="bold")
plt.grid(b=True, which='both', axis='both', color='lightgrey', linestyle='--', linewidth=1)
plt.legend()
plt.show()
```
